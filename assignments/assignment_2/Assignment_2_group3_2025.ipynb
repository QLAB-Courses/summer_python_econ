5. We want to calculate a series of new variables( or new columns) :  

In the summary : 
- We want the number of household dependent. These are the ones that don't percieve any income or the difference between the total household  members  and the total of percievers. 
- We want the average monthly income from the household percievers. 
- We want the difference between income and expenditure of the houselholds.

````python
# I create a copy of module34_summary to ensure the DataFrame is independent
module34_summary = module34_summary.copy()

#I calculate the number of household dependents as the difference between total members and income earners

module34_summary.loc[:, "dependents"] = module34_summary["mieperho"] - module34_summary["percepho"]

#I calculate the average monthly income for income earners in the household
module34_summary.loc[:, "monthly_income"] = module34_summary.apply(
    lambda row: (row["inghog2d"] / 12) / row["percepho"] if row["percepho"] > 0 else 0,
    axis=1
)
# I calculate the difference between income and expenditure for each household
module34_summary.loc[:, "income_expenditure_diff"] = module34_summary["inghog2d"] - module34_summary["gashog2d"]

# Finally, confirm the results
print("Module 34 (summary) with the new variables:")
print(module34_summary[["dependents", "monthly_income", "income_expenditure_diff"]].head())

````

6. We want the `poverty` variable in the summary to be codified as an integers 1, 2 ,3 according to:
`1`: extreme poor home
`2`: not extreme poor home
`3`: not poor home  

Using this variable , make two more: 
1. A variable that says  `'extreme poor'`, `' not extreme poor'` and `'not poor '` instead of 1,2,3.
2. A variable that has `'poor'` and  `'not poor'` (or one that combines the poor and not extreme poor households).

````python
#1.6.1
#I start by recoding the "pobreza" variable into descriptive labels.
#This will help me better understand the data by assigning textual labels to numeric codes.
module34_summary.loc[:, "poverty_label"] = module34_summary["pobreza"].map({
    1: "extreme poor",
    2: "not extreme poor",
    3: "not poor"
})

#Next, I create a new column to group households into two categories: "poor" and "not poor."
# This combines the "extreme poor" and "not extreme poor" into a single "poor" group.
module34_summary.loc[:, "poverty_group"] = module34_summary["pobreza"].apply(
    lambda x: "poor" if x in [1, 2] else "not poor"
)

#Finally, I confirm the results by displaying a preview of the new variables
print("Preview of poverty-related variables in module34_summary:")
print(module34_summary[["pobreza", "poverty_label", "poverty_group"]].head())
````

````python
module34_summary
````

````python
#1.6.2
# First, I ensure that the NBI columns ("nbi1", "nbi2", "nbi3", "nbi4", "nbi5") are numeric.
# If there are any non-numeric values, I coerce them to NaN and then replace them with 0.
nbi_columns = ["nbi1", "nbi2", "nbi3", "nbi4", "nbi5"]
module1_1_code[nbi_columns] = module1_1_code[nbi_columns].apply(pd.to_numeric, errors="coerce").fillna(0)


# Next, I create a new column called "poverty_status" to classify households based on their basic needs (NBI).
# If a household has at least one unmet basic need (sum > 0), it is classified as "poor."
# Otherwise, it is classified as "not poor."
module1_1_code.loc[:, "poverty_status"] = module1_1_code[nbi_columns].sum(axis=1).apply(
    lambda x: "poor" if x > 0 else "not poor"
)

#Finally, I confirm the results by displaying a preview of the new variable "poverty_status."
print("Preview of poverty_status in module1_1_code:")
print(module1_1_code[["nbi1", "nbi2", "nbi3", "nbi4", "nbi5", "poverty_status"]].head())
````

7. We want to now the average household income (inghog2d) per poverty level. Make a `groupby` to find this.

(For example, the extreme poor households would only have one mean and it's the same for the others two  poor categories). 

````python
# I group the data by poverty level (poverty_label) and calculate the average household income (inghog2d)
average_income_by_poverty = module34_summary.groupby("poverty_label")["inghog2d"].mean()

#  I confirm the results by printing the average household income for each poverty level
print("Average household income by poverty level:")
print(average_income_by_poverty)
````

8. We whould now merge the data  
But before a sanity check: Verify that the number of rows in both data bases are the same.( remember we filtrate the observations  by  `result`).  The ENAHO  recolects the information of these 34,490 homes trough all their modules. In order to use all the modules at once, we would merge them.

````python
# First, I verify the number of rows in both datasets to ensure they match
print("Number of rows in module34_summary:", len(module34_summary))
print("Number of rows in module1_1_code:", len(module1_1_code))

# Now, I merge both datasets using an inner join based on the unique household identifiers
merged_data = pd.merge(
    module34_summary, 
    module1_1_code, 
    on=["conglome", "vivienda", "hogar"], 
    how="inner"
)

# Finally, I confirm the results of the merge by checking the number of rows and previewing the data
print("Number of rows after the merge:", len(merged_data))
print("Preview of merged data:")
print(merged_data.head())
````
