5. We want to calculate a series of new variables( or new columns) :  

In the summary : 
- We want the number of household dependent. These are the ones that don't percieve any income or the difference between the total household  members  and the total of percievers. 
- We want the average monthly income from the household percievers. 
- We want the difference between income and expenditure of the houselholds.

````python
# OBS: we're working with this df, which contains the variables of our interest
df_sum = df_sum.copy()
````

````python
#### For the nuber of household dependent 
df_sum["miedep"] = df_sum["mieperho"] - df_sum["percepho"] 

#### For the average monthly income from the household percievers

# There could be houses with no household percievers, which could lead a division by cero (NaN)
# For that reason we make sure to replace the Nan with 0
df_sum["inmenpro"] = (df_sum["inghog2d"]/12/df_sum["percepho"]).fillna(0)

#### For the difference between income and expenditure of the households

df_sum["difingas"] = df_sum["inghog2d"] - df_sum["gashog2d"]
````

6. We want the `poverty` variable in the summary to be codified as an integers 1, 2 ,3 according to:
`1`: extreme poor home
`2`: not extreme poor home
`3`: not poor home  

Using this variable , make two more: 
1. A variable that says  `'extreme poor'`, `' not extreme poor'` and `'not poor '` instead of 1,2,3.
2. A variable that has `'poor'` and  `'not poor'` (or one that combines the poor and not extreme poor households).

````python
df_sum["pobreza"] = df_sum["pobreza"].astype(int)  # Convert the poverty variable to integers
pobreza_labels = {    # Create a new variable with descriptive labels for poverty levels
    1: "extreme poor",
    2: "not extreme poor",
    3: "not poor"
}
df_sum["pobreza_label"] = df_sum["pobreza"].map(pobreza_labels)
````

````python
# Create a new variable that combines poor and not poor households
df_sum["pobreza_comb"] = df_sum["pobreza"].apply(lambda x: "poor" if x in [1, 2] else "not poor")
df_sum
````

7. We want to now the average household income (inghog2d) per poverty level. Make a `groupby` to find this.

(For example, the extreme poor households would only have one mean and it's the same for the others two  poor categories). 

````python
# Group by poverty level and calculate the average household income
main_pobreza = df_sum.groupby("pobreza_label")["inghog2d"].mean().reset_index()
main_pobreza.columns = ["poverty_level", "average_income"]
main_pobreza
````
